```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_excel("satislar.xlsx") # verinin okunması
```

```python
veriler 
```

```python
aylar = veriler[["Aylar"]] # bağımsız değişkeni ele alma
aylar
```

```python
satislar = veriler[["Satislar"]] # bağımlı değişkeni ele alma
satislar
```

```python
from sklearn.model_selection import train_test_split # train ve test olarak ayırma işlmei için import edildi.
```

```python
x_train, x_test, y_train, y_test = train_test_split(aylar,satislar, test_size=0.33, random_state=0) # x=bağımsız, y=bağımlı
```

MODEL İNŞASI

```python
from sklearn.linear_model import LinearRegression # lineer regresyon modelini kullanmak için import edildi.
```

```python
lr = LinearRegression()
```

Aşağıda("(lr.fit(X_train, Y_train)") X_train ve Y_train'in mevcut durumu modele öğretilir. Yani "Benim verilerim bunlar, haberin olsun." denir. Aynı zamanda X_train ve Y_train birlikte verilerek arasındaki ilişki de öğretilmiş olur. Yani "X_train'e göre Y_train'im budur." denir.  

```python
lr.fit(x_train, y_train) 
```

MODELİN UYGULANMASI

```python
tahmin = lr.predict(x_test) # Y_test'in karşılığı. Bu aşamadan sonra X_test'ten'in tahmininden gelen sonuç Y_test ile karşılaştırılabilir. 
```

```python
tahmin ## Bu tahmin y için yani bağımlı değişken için yapılan tahmindir. Bu değer ile y_test karşılaştırılır.
```
