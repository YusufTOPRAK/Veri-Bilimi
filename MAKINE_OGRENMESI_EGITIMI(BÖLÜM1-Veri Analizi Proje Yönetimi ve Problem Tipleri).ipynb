Veri setindeki eksik verileri, ilgili kolondaki verilerin ortalamasını alarak eksik veriler yerine yerleştirmek(Sci-kit Learn Kütüphanesi ile)

```python
from sklearn.impute import SimpleImputer ## Eksik verileri doldurmak için kullanıldı
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') 
## SimpleImputer bir class olduğu için imputer nesnesine atandı.
## missing_values = Hangi verinin doldurulacağını belirler.
## strategy = Hangi metot ile doldurulacağını belirler.
```

```python
yas = veriler[["boy","kilo", "yas"]].values
type(yas)
```

```python
imputer = imputer.fit(yas[:,1:4])
yas[:,1:4] = imputer.transform(yas[:,1:4])
yas
```

# KONU2: KATEGORİK VERİLER

Bu derste verilerin sayısal verilere dönüşümü gösterilecektir.

Cinsiyet özelliği sadece kadın ve erketen oluştuğu için rahat bir şekilde kadın yerine 1 erkek yerine 0 verilerek bu değerler sayısal verilere çevrilebirlir. Fakat ülke gibi birden fazla şıkka sahip özellikler için bu yapılamaz. Bunun için birden fazla şıkka sahip olan özellikler kolon olarak ele alınır ve her satır için 0 ve 1 değerleri verilir. Hangi kolonlor 0 ise o satırda o ülkeler yok, hangi kolon 1 ise o satırda 1 olan ülke var anlamına gelir. 

NOT: veriler.iloc[:, 0:1] ==> Çıktı DataFrame olarak döner. Seçilen sütunları içeren DataFrame döndürür. veriler.iloc[:,0] ==> Çıktı Series olarak döner. Yalnızca sütunun değerlerini içeren bir seri döndürür.

```python
ulkeler = veriler.iloc[:, 0:1].values ## iloc ==> :== satıların hepsini seçti ","'den sonra kolon seçimi 0:1=ilk kolon
```

```python
ulkeler
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder() ## nominal verileri sayısal verilere çevirir.
```

```python
ulkeler[:, 0] = le.fit_transform(veriler.iloc[:,0])
```

```python
ulkeler
```

```python
ohe = preprocessing.OneHotEncoder() ## Sayısal veriye çevrilen nominal verileri sütun haline getirir.
```

```python
ulkeler = ohe.fit_transform(ulkeler).toarray() ## toarray() = sonucu numpyarray şeklinde almak için kullanıldı.
```

```python
ulkeler
```

# KONU3: VERİ KÜMESİNİN EĞİTİM VE TEST OLARAK BÖLÜNMESİ  

Bu bölme işlemi, modelin eğitildiği veri üzerinde değil, ayrı bir veri seti üzerinde test edildiği için modelin gerçek dünya verilerine ne kadar iyi uyarlandığını değerlendirmek için kullanılır. Eğitim ve test veri setlerini ayrı tutmak, modelin aşırı uyum (overfitting) sorunlarını tespit etme ve genelleme yeteneğini değerlendirme açısından önemlidir.

```python
from sklearn.model_selection import train_test_split
```

Veriler bağımsız değişken(s(cinsiyet dışındakiler)) ve bağımlı değişken(sonuc3(cinsiyet)) olarak ayrılır. test_size ile bağımlı ve bağımsız değişkenlerin yüzde kaçının test kısmına ayrılacağı belirlenir. Geriye kalan kısım değişkenlerin train kısmına ayrılır. random_state ile verilerin bu bölünmeye rastgele dağılması sağlanır.

```python
x_train, x_test, y_train, y_test =  train_test_split(s,sonuc3, test_size=0.33, random_state=0)
```
