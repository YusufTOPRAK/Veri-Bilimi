# DEĞER SEÇİMİ-GERİ ELEME(BACKWARD ELIMINATION)
1. 1.adımda geri eleme için bir sistem oluşturulmalı ve bu sistemde verilerin değişkenleri ve sabit değer olmalı.
2. 2.adımda üzerinde çalışacağımız verilerin değerleri ele alınır. Sonrasında array içine atılır.
3. 3.adım OLS ile model oluşturulur. OLS'ye verilecek ilk parametre sonuç olarak bulunmak istenen değer yani bağımlı değişken(boy), ikinci parametre ise elemek için üzerinde test yaptığımız değerler yani bağımsız değişkenlerdir(X_l).
4. 4.adımda oluşturulan modelin çıktısını görmek için "summary()" metodu kullanılır ve bu değerlere göre hareket edilir.
5. 5.adımda P değerine göre eleme yapılmaya başlanır. P değeri ne kadar düşükse o kadar iyidir.

```python
import statsmodels.api as sm # modelin ve modelin değişkenlerinin başarısını ölçmek için kullanıldı.
```

```python
veri.head()
```

1.Adım: Geri eleme üzerinde çalışılacak sistem oluşturma. Burada üzerinde eleme yapacağımız verinin değişkenleri yer alır. Sabit değer verinin içinde olmadığı için "np.ones((22,1)).astype(int)" koduyla sabit değer eklenir(22 satır 1 kolon, türü = int). Sabit değer çarpanları 1 olduğu için 1 olarak belirlendi(anlamadım).

```python
X = np.append(arr = np.ones((22,1)).astype(int), values = veri, axis=1)
```

2.Adım: Üzerinde çalışacağımız verilerin değerleri ele alınır. Sonrasında array içine atılır.

```python
X
```

```python
X_l = veri.iloc[:, [0,1,2,3,4,5]].values
```

```python
X_l = np.array(X_l, dtype = float)
```

3.Adım: OLS metoduna ilgili parametreler(boy, X_l) verilir ve fit edilir.

```python
model = sm.OLS(boy, X_l).fit()
```

4.Adım: Oluşturulan modelin "model.summary()" koduyla çıktısı alınır ve bu değerlere göre hareket edilir.

```python
model.summary() # oluşturulan modelin raporu
```

NOT: P değeri ne kadar düşükse o kadar iyi

5.Adım: P değerlerine bakılarak eleme işlemine başlanır.

Alınan sonuçlarda 5. değişken yani 4.index(yas)'in P değeri yüksek çıktığı için elendi. Ve model tekrar çalıştırıldı. Ve yeni değerler elde edilidi.

```python
X_l = veri.iloc[:, [0,1,2,3,5]].values
X_l = np.array(X_l, dtype = float)
model = sm.OLS(boy, X_l).fit()
```

```python
model.summary()
```
